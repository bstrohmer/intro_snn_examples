# Izhikevich Example

Before creating the neural network, we need to select a simulator. In this case, we will use the Neural Simulation Tool (NEST). The first step is to import the relevant libraries / modules.

```python
import nest
import nest.voltage_trace
import matplotlib.pyplot as plt

nest.set_verbosity("M_WARNING")
nest.ResetKernel()
```

Next, we can create a neuron using a pre-defined neuron model in NEST. For more information about pre-defined neuron models and parameters that can be set for each, go to <https://nest-simulator.readthedocs.io/en/v2.18.0/models/neurons.html>.  
For this example, we will use the Izhikevich neuron. Additionally, we would like to see the activity of the neuron so we will create a voltmeter.

```python
neuron = nest.Create("izhikevich")
voltmeter = nest.Create("voltmeter")
```

To induce neuronal activity, we supply a current bias or current input to the neuron using the parameter I_e. 

```python
neuron.I_e = 400.0
```

Next, we can connect the voltmeter to the neuron to monitor activity.

```python
nest.Connect(voltmeter, neuron)
```

Now we can run the simulation.

```python
nest.Simulate(1000.0)
```

Finally, we can visualize what the voltmeter recorded by reading from it and plotting it.

```python
nest.voltage_trace.from_device(voltmeter)
plt.show()
```

Instead of re-running each of the code blocks individually, we can put them together in a single block and begin testing our parameters.  
What would happen if we changed the amount of input current? Can you change the parameters of the plot to zoom in on output activity? 

```python
import nest
import nest.voltage_trace
import matplotlib.pyplot as plt

nest.set_verbosity("M_WARNING")
nest.ResetKernel()

neuron = nest.Create("izhikevich")
voltmeter = nest.Create("voltmeter")

neuron.I_e = 400.0  #Update this value
nest.Connect(voltmeter, neuron)
nest.Simulate(1000.0)

nest.voltage_trace.from_device(voltmeter)
plt.show()
```
